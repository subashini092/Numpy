## 5. Array Reshaping and Flattening

```python
import numpy as np

arr = np.array([1, 2, 3, 4, 5, 6])

reshaped = arr.reshape(2, 3)

print("Original Array:")
print(arr)

print("\nReshaped Array:")
print(reshaped)
```

```python
import numpy as np

arr = np.array([[10, 20, 30],
                [40, 50, 60]])

flat = arr.flatten()

print("Original Array:")
print(arr)

print("\nFlattened Array:")
print(flat)
```

## 6. Array Stacking and Splitting

```python
a = np.array([1,2,3])
b = np.array([4,5,6])

print(np.vstack((a,b))) # vertical stacking - row wise

print(np.stack((a,b))) # horizontal stacking - column wise
```

```python
c = np.array([[1,2,3], [4,5,6]])
print(c)
```

split = np.hsplit(c,3) # horizontal split
for s in split:
    print(s)

```python
split = np.vsplit(c,2) # vertical split
for s in split:
    print(s)
```
